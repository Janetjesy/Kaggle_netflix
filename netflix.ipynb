#### duration: Even though it's an object, it contains numbers (e.g., "90 min" or "1 Season"). You can't calculate the average movie length until you strip the text and convert it to an integer.

#### date_added: This is a string right now (e.g., "September 24, 2021"). To find out if Netflix added more movies in the summer vs. winter, you'll eventually want to convert this to a datetime object.

## DATA CLEANING 

### First i create a copy of the file so if i mess it does not affect the main file

```python
netflix_copy=df.copy()
print("Copy of the file has been created")
```

```python
# Confirming it worked
print("Netflix Copy created!")
print(f"Original rows: {df.shape[0]}")
print(f"Copy rows: {netflix_copy.shape[0]}")
```

```python
##filling the missing values in director instead of removing the 2000 plus rows
df['director']=df['director'].fillna('Unkown')
```

```python

netflix_copy['director'] = netflix_copy['director'].fillna('Unknown')
netflix_copy['cast'] = netflix_copy['cast'].fillna('Unknown cast')
netflix_copy['country'] = netflix_copy['country'].fillna('Unknown country')
```

```python
netflix_copy.isnull().sum()
```

```python
##droping rows missing values
netflix_copy.dropna(subset=['rating', 'duration', 'date_added'], inplace=True)
```

```python
netflix_copy.isnull().sum()
```

```python
netflix_copy.shape
```

to solve the durat

```python
# Create a column that just has the number
netflix_copy['duration_num'] = netflix_copy['duration'].str.split(' ').str[0].astype(int)

# Create a column that tells us the unit (min or Season)
netflix_copy['duration_unit'] = netflix_copy['duration'].str.split(' ').str[1]
```

```python
print("--- Cleaning Complete ---")
print(f"Original Dataset Rows: {df.shape[0]}")
print(f"Cleaned Copy Rows: {netflix_copy.shape[0]}")
print(f"Rows removed (the tiny errors): {df.shape[0] - netflix_copy.shape[0]}")
print("\nMissing Values in netflix_copy:")
print(netflix_copy.isnull().sum())
```

```python
# Selecting all columns with objects
text_columns = netflix_copy.select_dtypes(include=['object']).columns

# Looping through them and removing text spaces
for col in text_columns:
    netflix_copy[col] = netflix_copy[col].str.strip()

print("Hidden spaces removed from all text columns!")
```

```python
 dropping useless columns
columns_to_drop = ['show_id'] 
netflix_copy.drop(columns=columns_to_drop, inplace=True)
```
