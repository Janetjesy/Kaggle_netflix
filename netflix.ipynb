## DATA CLEANING 

### First i create a copy of the file so if i mess it does not affect the main file

```python
netflix_copy=df.copy()
print("Copy of the file has been created")
```

```python
# Confirming it worked
print("Netflix Copy created!")
print(f"Original rows: {df.shape[0]}")
print(f"Copy rows: {netflix_copy.shape[0]}")
```

```python
##filling the missing values in director instead of removing the 2000 plus rows
df['director']=df['director'].fillna('Unkown')
```

```python
#filling missing values
netflix_copy['director'] = netflix_copy['director'].fillna('Unknown')
netflix_copy['cast'] = netflix_copy['cast'].fillna('Unknown cast')
netflix_copy['country'] = netflix_copy['country'].fillna('Unknown country')
```

```python
netflix_copy.isnull().sum()
```

```python
##droping rows missing values
netflix_copy.dropna(subset=['rating', 'duration', 'date_added'], inplace=True)
```

```python
netflix_copy.isnull().sum()
```

```python
netflix_copy.shape
```

## to solve the duration ...splitting the number from the string 

```python
# Create a column that just has the number
netflix_copy['duration_num'] = netflix_copy['duration'].str.split(' ').str[0].astype(int)

# Create a column that tells us the unit (min or Season)
netflix_copy['duration_unit'] = netflix_copy['duration'].str.split(' ').str[1]
```

## splitting the year from the string
##### By doing this one can actaully serach from the model eg what year were more movies created


```python
# Convert text dates to actual Python dates
netflix_copy['date_added'] = pd.to_datetime(netflix_copy['date_added'].str.strip())

# Now you can easily pull out the year or month!
netflix_copy['year_added'] = netflix_copy['date_added'].dt.year
```

```python
print("--- Cleaning Complete ---")
print(f"Original Dataset Rows: {df.shape[0]}")
print(f"Cleaned Copy Rows: {netflix_copy.shape[0]}")
print(f"Rows removed (the tiny errors): {df.shape[0] - netflix_copy.shape[0]}")
print("\nMissing Values in netflix_copy:")
print(netflix_copy.isnull().sum())
```

### Cast Count Feature

### Creating a new column called `cast_count` to represent the number of actors in each movie or TV show.

###  If the cast information is missing or labeled as "Unknown Cast", the value is set to 0.
###  Otherwise, the cast string is split using commas, and the number of actors is counted.

### This helps in analyzing how cast size varies across different titles.

```python
# Count how many actors are in the 'cast' string
# We split by comma and count the items
netflix_copy['cast_count'] = netflix_copy['cast'].apply(lambda x: 0 if x == 'Unknown Cast' else len(x.split(',')))
```

## The primary country feature 
## some movies or shows have more than one country listed in it this may be difficult to graph so let us recognize only the firts country mentioned.

```python
# Take only the first country before the comma
netflix_copy['primary_country'] = netflix_copy['country'].str.split(',').str[0]
```

# data standardization
### Space and case cleanup

```python
# Select all columns that are "object" (text) types
text_columns = netflix_copy.select_dtypes(include=['object']).columns

# Loop through them and remove any hidden spaces
for col in text_columns:
    netflix_copy[col] = netflix_copy[col].str.strip()

print("Hidden spaces removed from all text columns!")
```

```python
netflix_copy['rating'] = netflix_copy['rating'].str.upper()
```

# UNIVARIATE ANALYSIS

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a professional style
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))

# Create the count plot
ax = sns.countplot(data=netflix_copy, x='type', palette='viridis')

# Add titles and labels
plt.title('Distribution of Content: Movies vs TV Shows', fontsize=15)
plt.xlabel('Content Type', fontsize=12)
plt.ylabel('Total Count', fontsize=12)

# Show the exact numbers on top of the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + 0.3, p.get_height() + 50))

plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Order the ratings by most common to least common
rating_order = netflix_copy['rating'].value_counts().index

sns.countplot(data=netflix_copy, y='rating', order=rating_order, palette='magma')

plt.title('Content Ratings Distribution (Most to Least Frequent)', fontsize=15)
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Rating Category', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10, 5))

# We use a Histogram for numbers like years
sns.histplot(netflix_copy['release_year'], bins=30, kde=True, color='royalblue')

plt.title('Content Release Year Distribution', fontsize=15)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Get the Top 10 countries only
top_10_countries = netflix_copy['primary_country'].value_counts().head(10)

sns.barplot(x=top_10_countries.values, y=top_10_countries.index, palette='rocket')

plt.title('Top 10 Content Producing Countries', fontsize=15)
plt.xlabel('Total Titles', fontsize=12)
plt.ylabel('Country', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Filtering for Movies only to get a clean 'minutes' distribution
movie_durations = netflix_copy[netflix_copy['type'] == 'Movie']['duration_num']

sns.histplot(movie_durations, bins=40, kde=True, color='crimson')

plt.title('Distribution of Movie Durations (in Minutes)', fontsize=15)
plt.xlabel('Duration (Minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Mark the average (mean) duration
plt.axvline(movie_durations.mean(), color='black', linestyle='--', label=f'Average: {movie_durations.mean():.1f} min')
plt.legend()

plt.show()
```
